```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 model
# Options: yolov8n (nano), yolov8s (small), yolov8m (medium), yolov8l (large), yolov8x (extra large)
model = YOLO('yolov8m.pt')

# Define where to save the training results IN YOUR DRIVE.
# This prevents you from losing your trained weights if Colab times out.
project_dir = '../Football_Training'
name_dir = 'run_1'

# Train the model
results = model.train(
    data='../football-players-detection.v20-rf-detr-m.yolov8/data.yaml',
    epochs=50,             # Start with 50-100. The model will stop early if it plateaus.
    imgsz=640,             # Standard image resolution for YOLO
    batch=16,              # Reduce to 8 if you encounter "CUDA Out of Memory" errors
    patience=10,           # Early stopping: stop training if no improvement for 10 epochs
    project=project_dir,   # Save path
    name=name_dir,         # Save folder name
    device=0,              # Force GPU usage
    workers=8              # Number of CPU workers loading data
)
```

```python
# Validate the model using the best weights saved from training
metrics = model.val()

print(f"mAP50: {metrics.box.map50:.4f}")
print(f"mAP50-95: {metrics.box.map:.4f}")
```
